## üèè Most runs scored against an opponent team by players.

```python
def run_scored(batsman_name):
    bat=ipl[ipl['batsman'] == batsman_name]
    return bat.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def plot_runs_scored(batsman_name):
    runs_ply = run_scored(batsman_name)
    runs_ply.plot(kind='bar')
```

```python
run_scored('MS Dhoni')
```

```python
plot_runs_scored('MS Dhoni')
```

```python
run_scored('V Kohli')
```

```python
plot_runs_scored('V Kohli')
```

* ### find the most destructive death over batsman in the history of ipl

* on the basis of **stirke rate** ,min play **200 balls** in death over (16-20)
* ** strike rate = (number of runs/number of balls)*100 **


```python
mask =ipl['over']> 15
deth_over=ipl[mask]    ## sotre the deth over 
```

```python
deth_over
```

```python
deth_bat=deth_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = deth_bat > 200  ## boolian indexcies
```

```python
bat200_list=deth_bat[x].index.tolist()  ## find the batsman which play 200 or > 200 ball
```

```python
# runs scored by all these batsmans
# ball played by these 43 batsmans
mask = deth_over['batsman'].isin(bat200_list)
play200=deth_over [mask]
```

```python
play200   ## which plyare who play 200 or more ball in deth over
```

```python
runs_deth = play200.groupby('batsman')['batsman_runs'].sum()
ball_deth = play200.groupby('batsman')['batsman_runs'].count()

strike_rate = (runs_deth / ball_deth) *100
```

```python
strike_rate.sort_values(ascending=False).head(10)
```

```python
strike_rate.sort_values(ascending=False).head(20).plot(kind='barh')
```

**merge cell**
* ## find **orange cap , purple cap holder** batsman 

```python
new_ipl=ipl.merge(match,left_on='match_id',right_on='id')
```

```python
ipl.shape
```

```python
match.shape
```

```python
season_run= new_ipl.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
```

```python
sortseason_run= season_run.sort_values(['season','batsman_runs'],ascending=[True,False])
```

```python
sortseason_run
```

```python
orange_cap = sortseason_run.drop_duplicates('season')
```

* ##### orange cap holder

```python
orange_cap
```

#### Ploting the graph

```python

df = pd.DataFrame(orange_cap)
seasons = sorted(df['season'].unique())

plt.figure(figsize=(15, 8))


for batsman, group in df.groupby('batsman'):
   
    total_runs_per_season = [group[group['season'] == season]['batsman_runs'].sum() for season in seasons]
    
 
    plt.bar(seasons, total_runs_per_season, label=batsman)

plt.xlabel('Season')
plt.ylabel('Runs')
plt.title('Batsman vs. Runs for Each Season')
plt.legend()
plt.xticks(seasons, rotation=90)  
plt.show()
```

* ### Purple cap holder

```python
season_wiket= new_ipl.groupby(['season','bowler'])['player_dismissed'].count().reset_index()
```

```python
season_wiket
```

```python
sortseason_wik= season_wiket.sort_values(['season','player_dismissed'],ascending=[True,False])
```

```python
sortseason_wik.drop_duplicates('season')
```

```python
purple_cap = sortseason_wik.drop_duplicates('season')
```

```python
purple_cap
```

```python
df = pd.DataFrame(purple_cap)
seasons = sorted(df['season'].unique())

plt.figure(figsize=(15, 8))


for bowler, group in df.groupby('bowler'):
   
    total_wikit_per_season = [group[group['season'] == season]['player_dismissed'].sum() for season in seasons]
    
 
    plt.bar(seasons, total_wikit_per_season, label=bowler)

plt.xlabel('Season')
plt.ylabel('player_dismissed')
plt.title('Purple cap Holder')
plt.legend()
plt.xticks(seasons, rotation=90)  
plt.show()
```

* ## Indivisual player **hieghst scored in stadiums**

```python
new_ipl['batsman_runs'] = new_ipl['batsman_runs'].fillna(0)

# Create the pivot table
stat = new_ipl.pivot_table(index='batsman', columns='venue', values='batsman_runs', aggfunc='sum', margins=True, margins_name='Total')
```

```python
# row corresponding to the player
player_stats = stat.loc['G Gambhir'].sort_values(ascending=False)
```

```python
player_stats.reset_index()
```

```python
def player_stats(batsman_name):
    player_stats = stat.loc[batsman_name].sort_values(ascending=False).head(6)
    print(player_stats.reset_index())
    player_stats.plot(kind='bar',x='venue')
```
